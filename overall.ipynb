Openers


```python
df = pd.read_csv(r"C:\Users\srd9cxl\OneDrive - Colruyt Group NV\Documents\cricket dataset\dataset\all_teams.csv")
df_opener = df[df['Roles'].str.contains('Opener', na=False)]
df_opener
```

```python
cols=["Player","Bat_Avg","Bat_Innings","Bat_SR"]
df_opener_filtered = df_opener[cols]
df_opener_filtered
```

```python
df_opener_filtered=df_opener_filtered.dropna(subset=["Bat_Avg", "Bat_SR", "Bat_Innings"])
df_opener_filtered #mean cant be computedf with nans
```

```python
#z score is X-avg/ std dev
#zscore for bt avg
mean_avg=df_opener_filtered["Bat_Avg"].mean()
std_avg=df_opener_filtered["Bat_Avg"].std()
df_opener_filtered["Bat_Avg_z"] = (df_opener_filtered["Bat_Avg"] - mean_avg) / std_avg
```

```python
#zscore for strike rate
mean_sr=df_opener_filtered["Bat_SR"].mean()
std_sr=df_opener_filtered["Bat_SR"].std()
df_opener_filtered["Bat_SR_z"] = (df_opener_filtered["Bat_SR"] - mean_sr) / std_sr
```

```python
#zscore for innings
mean_innings=df_opener_filtered["Bat_Innings"].mean()
std_innings=df_opener_filtered["Bat_Innings"].std()
df_opener_filtered["Bat_Innings_z"] = (df_opener_filtered["Bat_Innings"] - mean_innings) / std_innings
```

```python
df_opener_filtered[["Bat_Avg_z", "Bat_SR_z", "Bat_Innings_z"]].mean() #should be close to 0 because of z score
```

```python
df_opener_filtered[["Bat_Avg_z", "Bat_SR_z", "Bat_Innings_z"]].std()#``
```

```python
#best opener based on SR
df_opener_filtered.sort_values(by=[ "Bat_SR_z"], ascending=False, inplace=True)
df_opener_filtered.head(10)
```

```python
#worst opener based on SR
df_opener_filtered.sort_values(by=[ "Bat_SR_z"], ascending=True, inplace=True)
df_opener_filtered.head(10)
```

```python
metrics_z = ["Bat_Avg_z", "Bat_SR_z", "Bat_Innings_z"]

variances = df_opener_filtered[metrics_z].var()
variances
#z score normalization results in variance of 1 for all metrics
#so we derive weights from variance of original data, then apply those weights to the z scores 
```

```python
raw_metrics = ["Bat_Avg", "Bat_SR"]

raw_variances = df_opener_filtered[raw_metrics].var()
raw_variances
```

```python
weights = raw_variances / raw_variances.sum()
weights
```

```python
#innings got high weight due to high variance but innings isnt a skill metric like avg and sr its a confience metric so lets try constraint it
```

```python
#lets cap innings contribution and treat it as a modifier
```

```python

raw_metrics = ["Bat_Avg", "Bat_SR"]

raw_variances = df_opener_filtered[raw_metrics].var()
raw_variances
```

```python
weights = raw_variances / raw_variances.sum()
weights
```

```python
df_opener_filtered["skill_score"] = (
    weights["Bat_Avg"]      * df_opener_filtered["Bat_Avg_z"] +
    weights["Bat_SR"]       * df_opener_filtered["Bat_SR_z"])
df_opener_filtered.sort_values("skill_score", ascending=False).head(10)
```

```python
#we can see not having innings is affecting the overall score significantly
#lets build a factor to multiply with the skill score using innings
import numpy as np

df_opener_filtered["innings_factor"] = np.log1p(df_opener_filtered["Bat_Innings"])
df_opener_filtered["innings_factor"] /= df_opener_filtered["innings_factor"].mean()
```

```python
df_opener_filtered["Opener_Score"] = (
    df_opener_filtered["skill_score"] * df_opener_filtered["innings_factor"]
)
```

```python
df_opener_filtered.sort_values("Opener_Score", ascending=False).head(10)
```

```python
##all rounder
# Economy    0.001942
# Bowl_SR    0.057863
# Bat_Avg    0.067366
# Bat_SR     0.872829

#zscore for economy(less is better)
df_all_rounders['Economy_zscore'] = -(df_all_rounders['BOWL_Econ'] - df_all_rounders['BOWL_Econ'].mean()) / df_all_rounders['BOWL_Econ'].std()
#z score for strike rate(less is better)
df_all_rounders['Bowl_SR_zscore'] = -(df_all_rounders['BOWL_SR'] - df_all_rounders['BOWL_SR'].mean()) / df_all_rounders['BOWL_SR'].std()
#zscore for bt avg
df_all_rounders['Bat_Avg_zscore'] = (df_all_rounders['BAT_Ave'] - df_all_rounders['BAT_Ave'].mean()) / df_all_rounders['BAT_Ave'].std()
#z score for bt sr
df_all_rounders['Bat_SR_zscore'] = (df_all_rounders['BAT_SR'] - df_all_rounders['BAT_SR'].mean()) / df_all_rounders['BAT_SR'].std()
#skill score without innings
df_all_rounders["skill_score_allrounder"] = (
    df_all_rounders["Economy_zscore"] * 0.002 +
    df_all_rounders["Bowl_SR_zscore"] * 0.058 +
    df_all_rounders["Bat_Avg_zscore"] * 0.067 +
    df_all_rounders["Bat_SR_zscore"] * 0.873
)
#innings factor
df_all_rounders["innings_factor_bowl"] = np.log1p(df_all_rounders["BOWL_Inns"])
df_all_rounders["innings_factor_bat"] = np.log1p(df_all_rounders["BAT_Inns"])
df_all_rounders["innings_factor_bowl"] /= df_all_rounders["innings_factor_bowl"].mean()
df_all_rounders["innings_factor_bat"] /= df_all_rounders["innings_factor_bat"].mean()
df_all_rounders["innings_factor"]= (df_all_rounders["innings_factor_bowl"] + df_all_rounders["innings_factor_bat"])/2
df_all_rounders["final_score_allrounder"] = df_all_rounders["skill_score_allrounder"] * df_all_rounders["innings_factor"]
df_all_rounders.sort_values("final_score_allrounder", ascending=False).head(10)
```
